```python
def calculate_average_ranks(metric_scores_list: List[np.ndarray]) -> np.ndarray:
    """
    Calculate average ranks for each model based on metric scores.
    
    Args:
    metric_scores_list (list of arrays): List of metric scores for each model.
    
    Returns:
    average_ranks (array): Average ranks for each model.
    """
    # Combine the metric scores for each variable across models
    all_metric_scores = np.vstack(metric_scores_list)
    
    # Calculate ranks for each variable across models
    ranks = rankdata(all_metric_scores, axis=0)
    
    # Calculate average ranks for each model
    average_ranks = np.mean(ranks, axis=1)
    
    return average_ranks
```
